## Step 3: Data Preparation

```python
## droping the columns which are not important- 
dataset.drop(['instant','dteday','casual','registered'],axis = 1,inplace = True)
dataset.head()
```

```python
### Converting numeric values to categorical data

dataset['mnth'] = dataset['mnth'].apply(lambda i: calendar.month_abbr[i])
```

```python
## mapping all seasons
dataset.season = dataset.season.map({1: 'Spring',2:'Summer',3:'Fall',4:'Winter'})
```

```python
## mapping weathersituations
dataset.weathersit = dataset.weathersit.map({1:'Clear',2:'Mist & Cloudy', 
                                             3:'Light Snow & Rain',4:'Heavy Snow & Rain'})
```

```python
## mapping days of week
dataset.weekday = dataset.weekday.map({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thrusday",5:"Friday",6:"Saturday"})
```

```python
# Checking the dataframe now

dataset.head()
```

### Dummy Variables

```python
### creating dummy variables for season , mnth ,weathersit ,weekday
dummy_variable = dataset[['season','mnth','weekday','weathersit']]
```

```python
dummy_variable = pd.get_dummies(dummy_variable,drop_first=True )
```

```python
## adding dummy variables to original dataset
dataset = pd.concat([dummy_variable,dataset],axis = 1)
```

```python
# Head of dataframe.
dataset.head()
```

```python
## droping the columns for which dummy variables were created
dataset.drop(['season', 'mnth', 'weekday','weathersit'], axis = 1, inplace = True)
```

```python
dataset.head()
```

```python
dataset.shape
```

## Step 4: Splitting the Dataset into Training and Testing Sets

Train-test split of dataset is done here.

```python
train, test = train_test_split(dataset, train_size = 0.7, test_size = 0.3, random_state = 100)
```

## Step 5 :Rescaling the Features 

```python
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the dummy variables.
scaling = ['cnt','hum','windspeed','temp','atemp']

train[scaling] = scaler.fit_transform(train[scaling])
```

```python
train.head()
```

```python
train.describe()
```

```python
# Checking the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (20, 10))
sns.heatmap(train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

Building a model using all the columns

### Dividing dataset into X and Y sets for the model building

```python
y_train = train.pop('cnt')
X_train = train
```

## Step 6: Building a linear model

Fit a regression line through the training data using `statsmodels`. In `statsmodels`, you need to explicitly fit a constant using `sm.add_constant(X)` because if we don't perform this step, `statsmodels` fits a regression line passing through the origin, by default.

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 13)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
#Columns selected by RFE and their weights
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

### Starting with all the columns selected by RFE

### Model

```python
#Print Columns selected by RFE. We will start with these columns for manual elimination
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
 
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Create a first fitted model
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
# Checking the parameters obtained

lm.params
```

```python
# Print a summary of the linear regression model obtained
print(lm.summary())
```

```python
## droping weekday_Saturday as its p > 0.05
```

```python
X_train_latest = X_train_rfe.drop(["weekday_Saturday"], axis = 1)
```

```python
##Rebuilding the model without 'weekday_Saturday'
```

```python
# Adding a constant variable 
X_train_lm = sm.add_constant(X_train_latest)
```

```python
lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model
```

```python
lm.summary()
```

### Checking VIF for multicollinearity

```python
# Calculate the VIFs for the new model
vif = pd.DataFrame()
X = X_train_latest
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### We want a VIF which is less than 5. So we need to drop some variables.

```python
X_train_latest = X_train_latest.drop(['const'], axis=1)
```

```python
# Calculating the VIFs for the new model again
vif = pd.DataFrame()
X = X_train_latest
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
### droping hum from the model
X_train_latest = X_train_latest.drop(['hum'], axis=1)
```

```python
# Adding a constant variable and rebuilding the model
X_train_lm = sm.add_constant(X_train_latest)

# Create a first fitted model
lm = sm.OLS(y_train,X_train_lm).fit()  
```

```python
# Calculating the VIFs for the new model again
vif = pd.DataFrame()
X = X_train_latest
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### This is the Final Model

## Step 7: Residual Analysis of the train data

Now we need to check if the error terms are normally distributed (which is a major assumption of linear regression). So we need to plot the histogram of the error terms to see if it holds true.

```python
y_train_cnt = lm.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Graph', fontsize = 20)                  # Plot heading 
plt.xlabel('Error', fontsize = 18)                         # X-label
```

## Step 8: Making Predictions Using the Final Model

Now that we have fitted the model and checked the normality of error terms, it's time to go ahead and make predictions using the final, i.e. third model.

#### Applying the scaling on the test sets

```python
num_vars = ['cnt','hum','windspeed','temp', 'atemp']


test[num_vars] = scaler.transform(test[num_vars])
```

```python
test.describe()
```

#### Dividing into X_test and y_test

```python
y_test = test.pop('cnt')
X_test = test
```

```python
# Adding constant variable to test dataframe
X_test = sm.add_constant(X_test)
```

```python
# predicting using values used by the final model
test_col = X_train_lm.columns
X_test=X_test[test_col[1:]]
# Adding constant to test dataframe
X_test = sm.add_constant(X_test)

X_test.info()
```

```python
# Making predictions using model

y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mse
```
